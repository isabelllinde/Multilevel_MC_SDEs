```python
import mlmc_RSCAM as mlmc
import numpy as np
import matplotlib.pyplot as plt
import inspect as inspect
plt.rc('text', usetex=True)
plt.rc('font',**{'serif':['cm']})
plt.style.use('seaborn-paper')
plt.rcParams.update({'axes.titlesize':36})
plt.rcParams.update({'axes.labelsize': 30})
plt.rcParams.update({'xtick.labelsize': 26})
plt.rcParams.update({'ytick.labelsize': 26})
plt.rcParams.update({'legend.fontsize': 30})
plt.rcParams.update({'legend.title_fontsize':30})
plt.rcParams.update({'figure.titlesize': 40})
plt.rcParams.update({'lines.linewidth' : 2})
```

```python
##American Option
def exb(opt,t):
    return np.maximum(opt.K*(1-0.8*np.sqrt(opt.T-t)),0)
Am=mlmc.Amer_GBM(exb)
#Make Giles-style mean,variance,num. levels and complexity plots
eps=[0.1,0.2,0.25,0.3]
fig,ax=plt.subplots(2,2,figsize=(40,24))
markers=['o','s','x','d']
mlmc.Giles_plot(Am,markers=markers,eps=eps,label='American GBM',fig=fig)
```
