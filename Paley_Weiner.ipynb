```python
import numpy as np
import matplotlib.pyplot as plt

def paley_weiner(t, M, r_0, r_n):
    A = r_0 * t/(np.sqrt(2*np.pi))
    n = np.arange(1, M+1)
    B = (2.0 / np.sqrt(np.pi)) * np.sum(np.divide(r_n * np.sin(0.5 * n * t), n))
    return A + B
```

```python
T = np.linspace(0, 5, 500)
# M = 5
# random variables - iid N(0,1); r_0 and random number vector (R^n) r_n
np.random.seed(0)

m_list = [1,2,5,10,50,200]
r_0 = np.random.randn(1)
r_n = np.random.randn(m_list[-1])

plt.figure(figsize=(15,12))

for m in m_list:
    # to get working - had to take first m values of r_n
    r_n_trun = r_n[:m]
    pw = []
    for t in T:
        pw.append(paley_weiner(t, m, r_0, r_n_trun))
    plt.plot(T, pw, label='M = {}'.format(m))
    
plt.xlabel('time')
plt.ylabel('W(t)')
plt.title('Truncated Paley-Weiner')
plt.legend()
plt.show()
```
